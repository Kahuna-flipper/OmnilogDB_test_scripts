```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
path1 = '/home/kkrishnan/SBRG/OmnilogDB/Data/DP Clinical Isolate/Metadata'
path2 = '/home/kkrishnan/SBRG/OmnilogDB/Data/CYS1_2020/Data'

onlyfiles = [f for f in listdir(path1) if isfile(join(path1, f))]

temp_data = pd.read_excel(path1+'/'+onlyfiles[0])
temp_data_2 = pd.read_excel(path1+'/'+onlyfiles[1])

metadata = pd.concat([temp_data,temp_data_2],ignore_index=True)

for i in range(2,len(onlyfiles)):
    temp_data = pd.read_excel(path1+'/'+onlyfiles[i])
    metadata = pd.concat([metadata,temp_data],ignore_index=True)

strains = metadata['Strain name'].unique().tolist()
species = metadata['Species'].unique().tolist()


plates = metadata['PM Type (PM1 or PM2, or PM3, or PM4...)']

for i in range(0,len(plates)):
    plates[i]=plates[i][0:2]+'0'+plates[i][2:]
    
metadata['PM Type (PM1 or PM2, or PM3, or PM4...)'] = plates

metadata
```

```python
onlyfiles = [f for f in listdir(path2) if isfile(join(path2, f))]


for sheet_index in range(0,len(onlyfiles)):
    
    temp_data = pd.read_csv(path2+'/'+onlyfiles[sheet_index])

    temp_data.rename(columns={'Field 1': 'Strain', 'Field 2': 'Media','Field 3':'Specie','Field 4':'Replicate'}, inplace=True)

    for i in range(0,temp_data.shape[0]):

        replicate = temp_data.iloc[i]['Strain'][-2:]

        for st in strains:
            if st in temp_data.iloc[i]['Strain']:
                strain = st

        specie = metadata[metadata['Strain name']==strain]['Species'].unique().tolist()[0]
        sample = metadata[metadata['Strain name'] == strain]

        plate_type = temp_data.iloc[i]['Plate Type']
        sample_plates = sample['PM Type (PM1 or PM2, or PM3, or PM4...)']

        for pl in sample_plates:
            if pl in plate_type:
                media = sample['Condition (Medium)']

        temp_data.at[i,'Strain'] = strain
        temp_data.at[i,'Media'] = media
        temp_data.at[i,'Specie'] = specie
        temp_data.at[i,'Replicate'] = replicate
        
        temp_data.to_csv(onlyfiles[sheet_index], index=False)
    

    
    
```
